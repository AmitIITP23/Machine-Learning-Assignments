```python
import numpy as np
import pandas as pd
import random
```

```python
K=4
```

```python
dataset=pd.read_csv('BR_mod.csv')
```

```python
clusters={}
for i in range(K):
    clusters[i+1]=[]
centroids={}
```

```python
dataset.fillna(method='ffill',limit=54,inplace= True)
dataset.fillna(method='bfill',limit=54,inplace= True)
dataset.fillna(method='ffill',limit=54,inplace= True)
dataset.fillna(method='bfill',limit=54,inplace= True)
dataset
```

```python
for i in range(K):
    centroids[i+1]=dataset.loc[random.randint(0,dataset.shape[0]-1),:].values
```

```python
for i in centroids:
    print(centroids[i])
```

```python
meanOfData=dataset.mean()
```

```python
for i in range(dataset.shape[0]):
    distFromCentroid1=np.sqrt(np.sum([(a-b)*(a-b) for a, b in zip(centroids[1], dataset.iloc[i,:].values)]))
    #print(distFromCentroid1)
    distFromCentroid2=np.sqrt(np.sum([(a-b)*(a-b) for a, b in zip(centroids[2], dataset.iloc[i,:].values)]))
    #print(distFromCentroid2)
    distFromCentroid3=np.sqrt(np.sum([(a-b)*(a-b) for a, b in zip(centroids[3], dataset.iloc[i,:].values)]))
    #print(distFromCentroid3)
    distFromCentroid4=np.sqrt(np.sum([(a-b)*(a-b) for a, b in zip(centroids[4], dataset.iloc[i,:].values)]))
    #print(distFromCentroid1)
    minDistance=min(distFromCentroid1,distFromCentroid2,distFromCentroid3,distFromCentroid4)
    if(minDistance==distFromCentroid1):
        clusters[1].append(dataset.iloc[i,:].values)
    elif(minDistance==distFromCentroid2):
        clusters[2].append(dataset.iloc[i,:].values)
    elif(minDistance==distFromCentroid3):
        clusters[3].append(dataset.iloc[i,:].values)
    elif(minDistance==distFromCentroid4):
        clusters[4].append(dataset.iloc[i,:].values)
```

```python
for i in range(100):
    df1=pd.DataFrame(clusters[1])
    df2=pd.DataFrame(clusters[2])
    df3=pd.DataFrame(clusters[3])
    df4=pd.DataFrame(clusters[4])
    
    meanOfCluster1=df1.mean()
    minDistanceFromCluster1= np.linalg.norm(meanOfCluster1 - clusters[1][0])
    for i in range(len(clusters[1])):
        distanceFromCluster1=np.linalg.norm(meanOfCluster1 - clusters[1][i])
        if(distanceFromCluster1<minDistanceFromCluster1):
            minDistanceFromCluster1=distanceFromCluster1
            centroids[1]=clusters[1][i]
            
    meanOfCluster2=df2.mean()
    minDistanceFromCluster2= np.linalg.norm(meanOfCluster2 - clusters[2][0])
    for i in range(len(clusters[2])):
        distanceFromCluster2=np.linalg.norm(meanOfCluster2 - clusters[2][i])
        if(distanceFromCluster2<minDistanceFromCluster2):
            minDistanceFromCluster2=distanceFromCluster2
            centroids[2]=clusters[2][i]
            
    meanOfCluster3=df3.mean()
    minDistanceFromCluster3= np.linalg.norm(meanOfCluster3 - clusters[3][0])
    for i in range(len(clusters[3])):
        distanceFromCluster3=np.linalg.norm(meanOfCluster3 - clusters[3][i])
        if(distanceFromCluster3<minDistanceFromCluster3):
            minDistanceFromCluster3=distanceFromCluster3
            centroids[3]=clusters[3][i]
            
    meanOfCluster4=df4.mean()
    minDistanceFromCluster4= np.linalg.norm(meanOfCluster4 - clusters[4][0])
    for i in range(len(clusters[4])):
        distanceFromCluster4=np.linalg.norm(meanOfCluster4 - clusters[4][i])
        if(distanceFromCluster4<minDistanceFromCluster4):
            minDistanceFromCluster4=distanceFromCluster4
            centroids[4]=clusters[4][i]
            
    for i in range(K):
        clusters[i+1]=[]
        
    for i in range(dataset.shape[0]):
        distFromCentroid1=np.sqrt(np.sum([(a-b)*(a-b) for a, b in zip(centroids[1], dataset.iloc[i,:].values)]))
        distFromCentroid2=np.sqrt(np.sum([(a-b)*(a-b) for a, b in zip(centroids[2], dataset.iloc[i,:].values)]))
        distFromCentroid3=np.sqrt(np.sum([(a-b)*(a-b) for a, b in zip(centroids[3], dataset.iloc[i,:].values)]))
        distFromCentroid4=np.sqrt(np.sum([(a-b)*(a-b) for a, b in zip(centroids[4], dataset.iloc[i,:].values)]))
        minDistance=min(distFromCentroid1,distFromCentroid2,distFromCentroid3,distFromCentroid4)
    
        if(minDistance==distFromCentroid1):
            clusters[1].append(dataset.iloc[i,:].values)
        elif(minDistance==distFromCentroid2):
            clusters[2].append(dataset.iloc[i,:].values)
        elif(minDistance==distFromCentroid3):
            clusters[3].append(dataset.iloc[i,:].values)
        elif(minDistance==distFromCentroid4):
            clusters[4].append(dataset.iloc[i,:].values)
    
            
        
        
    
    
```
